## Anaconda 的命令行使用
1. conda -v 查看conda的版本
2. conda up date conda 更新conda版本
3. conda env list 查看所有的虚拟环境
4. conda create -n 'name' 'language(python=3.12)' 新建新的虚拟环境
5. conda activate 'name' 切换到对应虚拟环境
6. conda deactivate 退出当前虚拟环境
7. conda list 查看当前环境中安装的所有包
8. conda install 'package-name' 安装包及其依赖项目
9. conda search 'package-name' 搜索包的历史版本
10. conda uninstall 'package-name' 删除包及其依赖关系
