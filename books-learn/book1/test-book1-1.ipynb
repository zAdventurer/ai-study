```python
# 用嵌套列表构造矩阵
A = [[0,5],
 [3,4],
 [5,0]]
# 取出行向量
print(A[0])
print([A[0]])
print(A[1])
print([A[1]])
print(A[2])
print([A[2]])
# 取出列向量
print([row[0] for row in A])
print([[row[0]] for row in A])
print([row[1] for row in A])
print([[row[1]] for row in A])
```
