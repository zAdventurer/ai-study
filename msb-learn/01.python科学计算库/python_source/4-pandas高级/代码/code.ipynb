### 数据集成

#### 方式一concat

```python
import pandas as pd
import numpy as np
# df1 一班考试成绩
df1 = pd.DataFrame(data = np.random.randint(0,150,size = [10,3]),# 计算机科目的考试成绩
                  index = list('ABCDEFGHIJ'),# 行标签，用户
                  columns=['Python','Tensorflow','Keras']) # 考试科目

# df2 二班考试科目
df2 = pd.DataFrame(data = np.random.randint(0,150,size = [10,3]),# 计算机科目的考试成绩
                  index = list('KLMNOPQRST'),# 行标签，用户
                  columns=['Python','Tensorflow','Keras']) # 考试科目

# df3 增加了两个考试科目（一班）
df3 = pd.DataFrame(data = np.random.randint(0,150,size = (10,2)),
                  index = list('ABCDEFGHIJ'),
                  columns=['PyTorch','Paddle'])

display(df1,df2,df3)
```

一班和二班成绩合并

```python
# np.concatenate() 
# axis = 0表示 进行行合并
pd.concat([df1,df2],axis = 0)
```

一班科目增加，将原来的科目和增加的科目进行合并

```python
# 合并时，列增加 NaN  == not a number 空数据 404
pd.concat([df1,df3],axis = 1)
```

#### join SQL数据库风格插入合并  
merge 合并

```python
import pandas as pd
import numpy as np
# 表一中记录的是name和体重信息
df1 = pd.DataFrame(data = {'name':['softpo','Daniel','Brandon','Ella'],'weight':[70,55,75,65]})
# 表二中记录的是name和身高信息
df2 = pd.DataFrame(data = {'name':['softpo','Daniel','Brandon','Cindy'],'height':[172,170,170,166]})
df3 = pd.DataFrame(data = {'名字':['softpo','Daniel','Brandon','Cindy'],'height':[172,170,170,166]})

display(df1,df2,df3)
```

df1、df2进行合并，df1表示体重、df2表示身高

```python
pd.merge(df1,df2,) # 根据共同的属性，进行合并
# 共同的属性是name共同拥有的是：sotfpo、Daniel、Barandon
```

```python
# 指定了合并时，根据哪一列进行合并
pd.merge(df1,df3,left_on = 'name',right_on = '名字')
```

异常值

```python
df2 = pd.DataFrame(data = np.random.randn(10000,3)) # 正态分布数据
df2
# 3σ过滤异常值，σ即是标准差
# cond = (df2 > 3*df2.std()).any(axis = 1)
# index = df2[cond].index # 不满足条件的行索引
# df2.drop(labels=index,axis = 0) # 根据行索引，进行数据删除
```

```python
df2.mean()
```

```python
df2.std() # 标准差
```

$ > 3\sigma$ 异常值 $\sigma$ 表示标准差

```python
# 比较运算
# 异常值少数
cond = df2.abs() > 3 * df2.std()
cond_0 = cond[0]
df2[cond_0]
```

获取每一列的异常

```python
cond = df2.abs() > 3 * df2.std()
cond_0 = cond[0] # 默认[] 只能取列索引
cond_1 = cond[1]
cond_2 = cond[2]

# 逻辑或运算，只要有一个满足，返回True
cond_ = cond_0 | cond_1 | cond_2

df2[cond_]
```

```python
cond = df2.abs() > 3 * df2.std() # 计算异常值

# axis = 1计算每一行：只要一行中有一个TRUE，返回True
# True，表示异常值
cond_ = cond.any(axis = 1) # 只要有一个为真,返回True

df2[cond_]
```

#### map Series

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,10,size = (10,3)),
                  index = list('ABCDEFHIJK'),
                  columns=['Python','Tensorflow','Keras'])
df.iloc[4,2] = None # 空数据

df
```

```python
# 1、map批量元素改变，Series专有
df['Keras'].map({9:'Hello',5:'World',np.NaN:'AI'}) # 字典映射
```

```python
def convert(x):
    if x == 9:
        return 'Hello'
    elif x == 5:
        return 'World'
    elif x is np.NaN:
        return 'AI'
    else:
        return x
df['Keras'].map(convert)
```

#### apply元素改变。既支持 Series，也支持 DataFrame

```python
import numpy as np

import pandas as pd

df = pd.DataFrame(np.random.randint(0,100,size = (30,3)),columns=['Python','Math','En'])
df
```

对一列进行了转换，并且，进行了插入

```python
def convert(x):
    if x < 60:
        return '不及格' # 条件满足，直接执行下一个
    elif x < 80: # 是否小于80，条件满足，60 <= x < 80
        return '中等'
    else:
        return '优秀'
# 根据规则，进行数据转换
result = df['Python'].apply(convert)

index = list(df.columns).index('Python') + 1

# 插入
df.insert(loc = index,column = 'Python' + '等级',value = result)

df
```

多行转换

```python
def convert(x):
    if x < 60:
        return '不及格' # 条件满足，直接执行下一个
    elif x < 80: # 是否小于80，条件满足，60 <= x < 80
        return '中等'
    else:
        return '优秀'

for col in ['Python','Math','En']: # 已经字符换
    # 根据规则，进行数据转换
    result = df[col].apply(convert)

    # 插入位置的索引，使用变量表示
    index = list(df.columns).index(col) + 1

    # 插入
    df.insert(loc = index,column = col + '等级',value = result)

df
```
