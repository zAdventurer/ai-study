### 知识点概述

#### 字符串替换

```python
s1 = '我大学体能极限是1000米3分34秒！'
# 将我字替换为："派"。
# replace()函数中第一个是替换前的，第二个替换后的
s1 = s1.replace('我', '派')
print(s1)

# 将成绩转化为纯数字
s2 = "4'04"
s2 = s2.replace('\'', '.')
s2 = float(s2)
print(s2)
```

#### 报错演示

```python
score = ["4'03", "4'34", "4'29", "3'48", "3'58",4]

# for循环批量替换
for s in score:
    s = s.replace('\'', '.')
    print(s)
```

#### 报错修复

```python
score = ["4'03", "4'34", "4'29", "3'48", "3'58",4]

# for循环批量替换
for s in score:
    # 判断s是不是字符串类型数据，这是一个新知识点
    if isinstance(s, str):
        s = s.replace('\'', '.')
    print(s)
```

### 数据转换

#### 男生1000米跑成绩转换

#### 男生1000米跑成绩转换并赋值

```python
import pandas as pd
# 加载男生的体测成绩
df_boy = pd.read_excel('./体测成绩.xlsx',sheet_name=0)
# 空数据处理：没有参加体能测试，成绩为0
df_boy = df_boy.fillna(0)
def convert(s):
    if isinstance(s, str):
        s = float(s.replace('\'', '.'))
    return s
df_boy['男1000米跑'] = df_boy['男1000米跑'].map(convert)
# 保存数据，index = False表示不存储行索引
df_boy.to_excel('./男生体测成绩.xlsx', index = False)
```

#### 体侧成绩评分表数据转换

```python
# 评分标准
score = pd.read_excel('./体侧成绩评分表.xls',header = [0,1])
score
def convert(x):
    m,s = [int(i) for i in x.strip('"').split("'")]
    return m + s/100.0
# 男生考核标准转换
score.iloc[:,-4] = score.iloc[:,-4].apply(convert)
# 女生考核标准准换
score.iloc[:,-2] = score.iloc[:,-2].apply(convert)
score = score.reset_index()
score.to_excel('./体侧成绩评分表_处理.xlsx',header=[0,1])
```

```python
%%time
import pandas as pd
# 加载处理之后的男生体测成绩
df_boy = pd.read_excel('./男生体测成绩.xlsx')
# 加载成绩评分表
score = pd.read_excel('./体侧成绩评分表_处理.xlsx', header = [0,1])
# 定义转换方法
def convert(x):
    if x == 0: # 说明没有参加体能测试，分数为0分
        return 0
    for i in range(20): # 成绩划分20等级
        if x <= score['男1000米跑']['成绩'][i]:
            return score['男1000米跑']['分数'][i]
    return 0 # 说明跑的太慢了，分数为0分
df_boy['男1000米跑'].apply(convert)
```
