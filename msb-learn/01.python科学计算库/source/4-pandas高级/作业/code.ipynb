### 知识点概述

#### 字符串替换

```python
s1 = '我大学体能极限是1000米3分34秒！'
# 将我字替换为："派"。
# replace()函数中第一个是替换前的，第二个替换后的
s1 = s1.replace('我', '派')
print(s1)

# 将成绩转化为纯数字
s2 = "4'04"
s2 = s2.replace('\'', '.')
s2 = float(s2)
print(s2)
```

#### 报错演示

```python
score = ["4'03", "4'34", "4'29", "3'48", "3'58",4]

# for循环批量替换
for s in score:
    s = s.replace('\'', '.')
    print(s)
```

#### 报错修复

```python
score = ["4'03", "4'34", "4'29", "3'48", "3'58",4]

# for循环批量替换
for s in score:
    # 判断s是不是字符串类型数据，这是一个新知识点
    if isinstance(s, str):
        s = s.replace('\'', '.')
    print(s)
```

### 男生体测成绩分数转换 <font color = 'green'>3'35" ---> 95分</font>

#### 男生1000米跑成绩分数转换

```python
%%time
import pandas as pd
# 加载处理之后的男生体测成绩
df_boy = pd.read_excel('./男生体测成绩.xlsx')
# 加载成绩评分表
score = pd.read_excel('./体侧成绩评分表_处理.xlsx', header = [0,1])
# 定义转换方法
def convert(x):
    if x == 0: # 说明没有参加体能测试，分数为0分
        return 0
    for i in range(20): # 成绩划分20等级
        if x <= score['男1000米跑']['成绩'][i]:
            return score['男1000米跑']['分数'][i]
    return 0 # 说明跑的太慢了，分数为0分
df_boy['男1000米跑'].apply(convert)
```

#### 批量转换男生速度类成绩分数

```python
%%time
import numpy as np
import pandas as pd
df_boy = pd.read_excel('./男生体测成绩.xlsx')
score = pd.read_excel('./体侧成绩评分表_处理.xlsx', header = [0,1])
cols = ['男1000米跑', '男50米跑']
def convert(x, col):
    if x == 0: # 说明没有参加体能测试，分数为0分
        return 0
    for i in range(20): # 成绩划分20等级
        if x <= score[col]['成绩'][i]:
            return score[col]['分数'][i]
    return 0 # 说明跑的太慢了，分数为0分
for col in cols:
    s = df_boy[col].apply(convert, args = (col,))
    columns = df_boy.columns.to_list()
    index = columns.index(col) + 1
    df_boy.insert(loc = index, column = col + '分数', value = s)
df_boy.head()
```

#### 批量转换男生力量型成绩分数并保存

```python
%%time
def convert(x, col):
    for i in range(20): # 成绩划分20等级
        if x >= score[col]['成绩'][i]:
            return score[col]['分数'][i]
    return 0 # 说明跳远不达标，分数为0分
cols = ['男跳远', '男体前屈','男引体', '男肺活量']
for col in cols:
    s = df_boy[col].apply(convert, args = (col,))
    columns = df_boy.columns.to_list()
    index = columns.index(col) + 1
    df_boy.insert(loc = index, column = col + '分数', value = s)
display(df_boy.head())

df_boy.to_excel('./男生体测成绩-分数.xlsx', index = False)
```
